```python
import pandas as pd
```

```python
results = ""
with open("trec_results.txt", "r") as f:
    results = f.read()
```

```python
df = pd.DataFrame(columns={'query_number', 'map', 'recall'})
```

```python
for query in results.split("\nnum_ret"):
    query_metrics = query.split("\n")
    _map = None
    _num_rel = None
    _num_rel_ret = None
    query_num = None
    for element in query_metrics:
        if "map" in element:
            query_num = element.split("\t")[1]
            _map = element.split("\t")[2]
        elif "num_rel_ret" in element:
            num_rel_ret = element.split("\t")[2]
        elif "num_rel" in element:
            num_rel = element.split("\t")[2]
        
            
    if _map != None: 
        map_element = {
                    "query_number": query_num,
                    "map":_map,
                    "recall": float(num_rel_ret)/float(num_rel)
                }        
        df = df.append(map_element, ignore_index=True)
```

```python
df.sort_values("recall", ascending=False)
```
